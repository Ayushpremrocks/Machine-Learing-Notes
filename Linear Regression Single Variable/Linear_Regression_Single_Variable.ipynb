```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

Homeprices

```python
df = pd.read_csv("/content/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
x = 3300
reg.predict([[x]])
```

```python
m = reg.coef_
m
```

```python
c = reg.intercept_
c
```

```python
y = m*x + c
y
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

canada per capita income

```python
ap = pd.read_csv("/content/canada_per_capita_income.csv")
ap
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(ap.year,ap['per capita income (US$)'],color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(ap[['year']],ap['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(ap.year,ap['per capita income (US$)'],color='red',marker='+')
plt.plot(ap.year,reg.predict(ap[['year']]),color='blue')
```
